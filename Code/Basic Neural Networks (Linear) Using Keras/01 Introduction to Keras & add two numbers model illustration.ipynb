## Exercise 1
* Create a neural network model using Keras to add two numers

````python
import numpy as np
import tensorflow as tf
from tensorflow import keras
````

````python
# data preparation

# create numpy array with two arrays of 1000 random values
x= np.array([np.random.randint(low=2, high=500, size=1000), np.random.randint(low=2, high=500, size=1000)])

# add the two arrays and append it to previously defined numpy array and transpose it
x = np.append(x, [x[0] + x[1]]).reshape(3,1000).T

# convert numpy array to a tensor
x_tensor = tf.convert_to_tensor(x)
````

````python
# separate training inputs and training targets
x_train   = x_tensor[:,0:2]
x_targets = x_tensor[:,2:3]

# normalise the data
#from tensorflow.keras.layers import Normalization

#normalizer = Normalization(axis=-1)
#normalizer.adapt(x_tensor)

#normalized_data = normalizer(x_tensor)

# separate training inputs and training targets
#x_train   = normalized_data[:,0:2]
#x_targets = normalized_data[:,2:3]
````

````python
# model construction

# import models and layers modules from keras package
from keras import models
from keras import layers

# create a neural networks of eight neurons in first layer and one neuron in the second layer
add_two_model = models.Sequential()
add_two_model.add(layers.Dense(8, activation='relu', input_shape=(2,)))
add_two_model.add(layers.Dense(1, activation='linear', name = 'out'))
````

````python
# compile model using loss as MAE
add_two_model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mae'])

# check the summary of the model
add_two_model.summary()
````

````python
# train the model
add_two_model.fit(x_train, x_targets, batch_size=32, epochs = 100)
````

````python
# validating model
test_set = np.array([[3.3, 30], [4.5, 10.5], [11.1,10.1], [2.9,10.1],[5.9,9.1], [14,104]])
print(add_two_model.predict(test_set))
````
