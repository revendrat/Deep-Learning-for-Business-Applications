## Exercise 2
* Create a neural network model using Keras to multiply two matrices

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
# data preparation

#set random seed to generate same random numbers to repeat the results
import random
random.seed(365)

#create NumPy matrix A of size 2x2 filled with random integers between 0 and 100
matA=np.random.randint(0, 100, (2, 2))

#create NumPy matrix B of size 2x2 filled with random integers between 0 and 100
matB=np.random.randint(0, 100, (2, 2))

# perform matrix multiplication
matAB = np.matmul(matA,matB)

#flatten the values of matrix A into array
x_train = matA.flatten()

#flatten values of matrix B into array, and append with matrix A in training data
x_train = np.append(x_train, matB.flatten())

#flatten values of matrix AB into array for target of training data
x_targets = matAB.flatten()

# use a for loop to generate 100 matrices of A and B, multiply them and plug them into train and target sets 
for i in range(1000):
  matA=np.random.randint(0, 100, (2, 2))
  matB=np.random.randint(0, 100, (2, 2))
  matAB = np.matmul(matA,matB)
  x_train = np.append(x_train, np.array([matA.flatten(), matB.flatten()]))
  x_targets = np.append(x_targets, matAB.flatten())

print(x_train.size)
print(x_targets.size)

# reshapte train and test sets to make them suitable for building neural network model
x_train = x_train.reshape(1001,8)
x_targets = x_targets.reshape(1001,4)

# convert numpy array to a tensor
x_train_tensor = tf.convert_to_tensor(x_train)
x_targets_tensor = tf.convert_to_tensor(x_targets)
```

```python
print(x_train_tensor[0])
print(x_targets_tensor[0])
```

```python
# multiply and verify the results
41*23+75*96 # first element of matrix multiplication
```

```python
# Min Max Scaling 
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_train_tensor_scaled = mms.fit_transform(x_train_tensor)
mms_target = MinMaxScaler()
x_targets_tensor_scaled = mms_target.fit_transform(x_targets_tensor)
```

```python
x_train_tensor = tf.keras.utils.normalize(x_train_tensor)
x_targets_tensor = tf.keras.utils.normalize(x_targets_tensor)
print(x_train_tensor[0])
print(x_targets_tensor[0])
```

```python
# model construction

# import models and layers modules from keras package
from keras import models
from keras import layers

# create a neural networks of eight neurons in first layer and one neuron in the second layer
matmul_two_model = models.Sequential()
matmul_two_model.add(layers.Dense(16, activation='relu', input_shape=(8,)))
matmul_two_model.add(layers.Dense(16, activation='relu'))
matmul_two_model.add(layers.Dense(32, activation='relu'))
matmul_two_model.add(layers.Dense(64, activation='relu'))
# matmul_two_model.add(layers.Dense(128, activation='relu'))
# matmul_two_model.add(layers.Dense(128, activation='relu'))
# matmul_two_model.add(layers.Dense(256, activation='relu'))
# matmul_two_model.add(layers.Dense(256, activation='relu'))
# matmul_two_model.add(layers.Dense(128, activation='relu'))
# matmul_two_model.add(layers.Dense(64, activation='relu'))
# matmul_two_model.add(layers.Dense(32, activation='relu'))
# matmul_two_model.add(layers.Dense(16, activation='relu'))
# matmul_two_model.add(layers.Dense(4, activation='linear', name = 'out'))
matmul_two_model.add(layers.Dense(4, activation='relu', name = 'out'))
```

```python
# compile model using loss as MAE
matmul_two_model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mean_absolute_error'])


# check the summary of the model
matmul_two_model.summary()
```

```python
# train the model
# matmul_two_model.fit(x_train_tensor, x_targets_tensor, batch_size=16, epochs = 50)
matmul_two_model.fit(x_train_tensor_scaled, x_targets_tensor, epochs = 50)
```
