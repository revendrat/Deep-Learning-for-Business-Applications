```python
"""Implementation of XOR Gate Using Keras"""

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

targets = np.array([ #XNOR, #XOR
    [1, 0], 
    [0, 1], 
    [0, 1], 
    [1, 0]
])
```

```python
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(2,)))
model.add(Dense(8, activation='relu', input_shape=(16,)))
model.add(Dense(2, activation='softmax'))

model.compile(optimizer= 'adam', #'rmsprop', #'sdg' 
                loss='categorical_crossentropy',
                metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(inputs, targets,
                    epochs=10,
                    batch_size=1)
```

```python
#history.model
history.params
```

```python
predictions = model.predict(inputs)

print('Expected Output:  ', targets.argmax(axis=-1))
print('Predicted Output:  ', predictions.argmax(axis=-1))
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
